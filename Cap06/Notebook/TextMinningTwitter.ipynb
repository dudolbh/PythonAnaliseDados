## An√°lise de Dados com Pandas e Scikit-Learn

```python
## An√°lise de Dados com Pandas e Scikit-Learn
# criando um dataset com dados retornados do MongoDB
dataset = [{"created_at": item["created_at"], "text": item["text"],} for item in col.find()] #Utilizar list comprehnetion
```

```python
# Importando o m√≥dulo Pandas para trabalhar com datasets em Python
import pandas as pd
```

```python
# Criando um dataframe a partir do dataset 
df = pd.DataFrame(dataset) #Converte o dataset em um dataframe
# Imprimindo o dataframe
df
# Primeiro indice √© 0
```

```python
# Importando o m√≥dulo Scikit Learn
from sklearn.feature_extraction.text import CountVectorizer #Conta quantas vezes a palavra aparece em determinado dataset
```

```python
# Usando o m√©todo CountVectorizer para criar uma matriz de documentos
cv = CountVectorizer()
count_matrix = cv.fit_transform(df.text)
```

```python
# Contando o n√∫mero de ocorr√™ncias das principais palavras em nosso dataset
word_count = pd.DataFrame(cv.get_feature_names(), columns=["word"])
word_count["count"] = count_matrix.sum(axis=0).tolist()[0]
word_count = word_count.sort_values("count", ascending=False).reset_index(drop=True)
word_count[:50]
```
